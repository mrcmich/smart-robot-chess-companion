```python
model = YOLO('yolov8s.pt')
training_results = model.train(data='chess.yaml', epochs=50, batch=32, imgsz=640, save_period=-1, device='cuda:0', workers=8, optimizer='Adam', verbose=True, resume=False)
```

```python
model = YOLO("/content/runs/detect/train/weights/best.pt")
validation_results = model.val(data='chess.yaml', batch=32, imgsz=640, device='cuda:0', workers=8, verbose=True, save=True, save_txt=True, conf=0.5)
```
